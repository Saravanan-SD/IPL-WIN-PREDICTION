```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
matches= pd.read_csv('matches.csv')
deliveries= pd.read_csv('deliveries.csv')
matches.head()
```

```python
deliveries.head()
```

```python
matches.shape,deliveries.shape
```

```python

totalrun_df = deliveries.groupby(['match_id','inning']).sum()['total_runs'].reset_index()

totalrun_df
```

```python
totalrun_df = totalrun_df[totalrun_df['inning']==1]
totalrun_df['total_runs'] = totalrun_df['total_runs'].apply(lambda x:x+1)
totalrun_df
```

```python
matches.columns
```

```python
match_df= matches.merge(totalrun_df[['match_id','total_runs']],
                        left_on='id',right_on='match_id')

match_df
```

```python
match_df['team1'].unique()
```

```python
teams=[
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'
]
```

```python
match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1']=match_df['team1'].str.replace('Deccan Chargers','Delhi Capitals')
match_df['team1']=match_df['team1'].str.replace('Deccan Chargers','Delhi Capitals')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]

match_df['team1'].unique()
```

```python
match_df[match_df['dl_applied']==1].style.background_gradient(cmap='plasma')
```

```python
match_df=match_df[match_df['dl_applied']==0]
```

```python
match_df=match_df[['match_id','city','winner','total_runs']]

match_df
```

```python
delivery_df=match_df.merge(deliveries,on='match_id')
```

```python
delivery_df.head(3)
```

```python
delivery_df.columns
```

```python
delivery_df=delivery_df[delivery_df['inning']==2]
delivery_df.head()
```

```python
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].transform(pd.Series.cumsum)
```

```python
delivery_df.head()
```

```python
delivery_df['runs_left']=delivery_df['total_runs_x']-delivery_df['current_score']
delivery_df.head()
```

```python
delivery_df['balls_left']=126-(delivery_df['over']*6+delivery_df['ball'])
```

```python
list(delivery_df['player_dismissed'].unique())
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna("0")
delivery_df['player_dismissed']=delivery_df['player_dismissed'].apply(lambda x:x if x=='0' else "1")
delivery_df['player_dismissed']=delivery_df['player_dismissed'].astype('int')

delivery_df['player_dismissed'].unique()
```

```python
wickets=delivery_df.groupby('match_id')['player_dismissed'].transform(pd.Series.cumsum)
```

```python
delivery_df['wickets']=10-wickets
delivery_df
```

```python
delivery_df['cur_run_rate']=(delivery_df['current_score']*6)/(120-delivery_df['balls_left'])
delivery_df['req_run_rate']=(delivery_df['runs_left']*6)/(delivery_df['balls_left'])
```

```python
delivery_df.head()
```

```python
def winner(row):
    return 1 if row['batting_team']==row['winner'] else 0
```

```python
delivery_df['results']= delivery_df.apply(winner,axis=1)
delivery_df.head()
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left',
                        'balls_left','wickets','total_runs_x','cur_run_rate',
                        'req_run_rate','results']]

final_df.head()
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df =final_df.dropna()
final_df.isnull().sum()
```

```python
final_df=final_df[final_df['balls_left']!=0]
```

```python
data=final_df.copy()
y= data['results']
X= data.drop(['results'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.info()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn import metrics
```

```python
cf = ColumnTransformer([
    ('trf',OneHotEncoder(drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
```

```python
pipe= Pipeline(steps=[
    ('step1',cf),
    ('step2',LogisticRegression(solver='liblinear'))
])

pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
print(metrics.accuracy_score(y_test,y_pred))
```

```python
pipe.predict_proba(X_test)[10]
```

```python
pipe2= Pipeline(steps=[
    ('step1',cf),
    ('step2',RandomForestClassifier())
])

pipe2.fit(X_train,y_train)
print(metrics.accuracy_score(y_test,pipe2.predict(X_test)))
```
